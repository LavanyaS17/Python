```python
def display(board):
    print(board[0], '|' ,board[1], '|' ,board[2])
    print('--|---|--')
    print(board[3], '|' ,board[4], '|' ,board[5])
    print('--|---|--')
    print(board[6], '|' ,board[7], '|' ,board[8])
```

```python
board = ['1','2','3','4','5','6','7','8','9']
display(board)
```

```python
def check_win(p1,p2,board):
    if (board[0]==board[1]==board[2]==p1 or board[3]==board[4]==board[5]==p1 or board[6]==board[7]==board[8]==p1 #horizontal
     or board[0]==board[3]==board[6]==p1 or board[1]==board[4]==board[7]==p1 or board[2]==board[5]==board[8]==p1 #vertical
     or board[0]==board[4]==board[8]==p1 or board[2]==board[4]==board[6]==p1):                                   #diagonal
        return 1
        
    elif(board[0]==board[1]==board[2]==p2 or board[3]==board[4]==board[5]==p2 or board[6]==board[7]==board[8]==p2
     or board[0]==board[3]==board[6]==p2 or board[1]==board[4]==board[7]==p2 or board[2]==board[5]==board[8]==p2
     or board[0]==board[4]==board[8]==p2 or board[2]==board[4]==board[6]==p2):                                   
        return 0
        
    else:
        print('')
```

```python
ask = 'yes'
while ask == 'yes':
    
    board = ['1','2','3','4','5','6','7','8','9']
    p1 = ''
    p2 = ''
    while (p1 != 'X' and p1 != 'O'):
        p1 = input("Player 1 do you want to be X or O? ")
    if p1 == 'X':
        p2 = 'O'
        print("Player 2 is O")
    else:
        p2 = 'X'
        print("Player 2 is X")
        
    print("\nPress a number to place your mark in the respective position\n")
    display(board)
    print("\nPlayer 1 plays first")
    turn = 1
    w = ''
    tie = ''
    while not(w==0 or w==1 or tie==1):
        if turn==1:
            pos = input("Player 1 enter position: ")
            while (pos.isalpha() == True or pos==''):
                print("Please enter a number")
                pos = input("Player 1 enter position: ")    
            pos = int(pos)
            while board[pos-1]=='X' or board[pos-1]=='O':
                print("This position is taken, enter some other number")
                pos = int(input("Player 1 enter position: "))
            board[pos-1] = p1
            display(board)
            w = check_win(p1,p2,board)
            if w==1:
                print("Player 1 won the game!")
                break
            else:
                turn = 2
            if set(board)=={'O', 'X'}:
                tie = 1
                print("Its a tie!")
                break
        if turn==2:
            pos = input("Player 2 enter position: ")
            while (pos.isalpha() == True or pos==''):
                print("Please enter a number")
                pos = input("Player 2 enter position: ")  
            pos = int(pos)
            while board[pos-1]=='X' or board[pos-1]=='O':
                print("This position is taken, enter some other number")
                pos = int(input("Player 2 enter position: "))
            board[pos-1] = p2
            display(board)
            w = check_win(p1,p2,board)
            if w==0:
                print("Player 2 won the game!")
                break
            else:
                turn = 1
            if set(board)=={'O', 'X'}:
                tie = 1
                print("Its a tie!")
                break
        
    ask = input("Do you want to play another game?(yes/no) ").lower()
    while not(ask.lower()=='yes' or ask.lower()=='no'):
        ask = input("Do you want to play another game?(yes/no) ").lower()
    print("\n")
    if ask == 'yes' or 'y':
        continue
    if ask == 'no' or 'n':
        break
```
